```python
import pandas as pd
import argparse
import scorer


def evaluate_predictions(predictions_file):
    # Step 1: Load the data
    df = pd.read_csv(predictions_file)

    # Step 2: Extract gold labels and predictions
    gold_labels = df["True_Labels"].tolist()
    predictions = df["Initial_Predictions"].tolist()

    # Step 3: Evaluate predictions
    p, r, f1 = scorer.score(gold_labels, predictions, verbose=True)

    # Step 4: Print results
    print("\nPredictions evaluation:")
    print(f"Precision: {p:.3%}")
    print(f"Recall:    {r:.3%}")
    print(f"F1:        {f1:.3%}")
```

## TACERED

```python
## BERT BASE
evaluate_predictions("./Predictions/PLM_BERT-BASE_TACRED_TEST.csv")
print("\n\n")
evaluate_predictions("./Predictions/PLM_BERT-BASE_TACRED_DEV.csv")
```
